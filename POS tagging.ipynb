```python
import nltk
import numpy as np
```

```python
from nltk.corpus import brown
```

```python
corpus_tagged = brown.tagged_sents()
```

```python
###adding start of sentence to the corpus
corpus_tagged_final = []
for i in range(0, len(corpus_tagged)):
    temp = [('/s', '/s')] + corpus_tagged[i]
    corpus_tagged_final.append(temp)
```

```python
brown_tags = [tag for s in corpus_tagged_final for (word, tag) in s]
```

```python
###get all distinct tags
distinct_tags_pre = set(brown_tags)
```

```python
## converting set to list
distinct_tags = []
for i in distinct_tags_pre:
    distinct_tags.append(i)
```

Language Models

```python
##language Model for bigram, trigram, etc
def language_model(n, corpus):
    my_dict = dict()
    
    for sentence in corpus:
        for i in range(0, len(sentence) - (n - 1)):
            if n == 1:
                key = sentence[i][1]
            else:
                lst = sentence[i : i + n]
                temp = []
                for i in lst:
                    temp.append(i[1])
                key = tuple(temp)

            if key in my_dict:
                my_dict[key] += 1
            else:
                my_dict[key] = 1
                
    my_dict = dict(sorted(my_dict.items(), key = lambda x : x[1], reverse = True))
    return my_dict
```

```python
## language model for bigram with observations
def language_model_observation(n, corpus):
    my_dict = dict()
    
    for sentence in corpus:
        for i in range(0, len(sentence) - (n - 1)):
            lst = sentence[i : i + n - 1]
            temp = []
            for i in range(0, len(lst)):
                temp.append(lst[i][1])
            temp.append(lst[len(lst) - 1][0])
            key = tuple(temp)
            
            if key in my_dict:
                my_dict[key] += 1
            else:
                my_dict[key] = 1  
                
#     my_dict = dict(sorted(my_dict.items(), key = lambda x : x[1], reverse = True))
    return my_dict
```

Probabilities and viterbi

```python
def probability_transmission(trigram): ##trigram ('NN', 'IN', 'the')
    count_num = 0
    count_deno = 1
    bigram = trigram[0:2]
    if trigram in trigram_tagged:
        count_num = trigram_tagged[trigram]
    if bigram in bigram_tagged:
        count_deno = bigram_tagged[bigram]
    return count_num/count_deno
```

```python
def probability_emmision(trigram):
    count_num = 0
    count_deno = 1
    bigram = trigram[0:2]
    if trigram in bigram_obs_tagged:
        count_num = bigram_obs_tagged[trigram]
    if bigram in bigram_tagged:
        count_deno = bigram_tagged[bigram]
    return count_num/count_deno
```

```python
###viterbi algorithm
def viterbi(sequence):
    words = sequence.split(' ')
    P_t = []
    Prev_tag = []
    tag_t = []
    
    for i in range(0, len(words) + 1):
        P_t.append([])
        Prev_tag.append([])
        tag_t.append([])
    idx = distinct_tags.index("/s")   
    
    for i in range(0, len(distinct_tags)):
        P_t[0].append(1/472)                       ##give random later
        Prev_tag[0].append(idx)
        Prev_tag[1].append(idx)
        tag_t[0].append(i)
        
    for i in range(0, len(distinct_tags)):
        trigram = ("/s", distinct_tags[i], words[0])
        tag_t[1].append(i)
        P_t[1].append(P_t[0][i] * probability_emmision(trigram))  ###P(tag/tag_prev) * prob(word/tag)
    
#     P_t[1] = dict(sorted(P_t[1].items(), key = lambda x : x[1], reverse = True))

#     ##take top k tags from previous
    for m in range(1, len(words)):
        i = m + 1
        for j in range(0, len(distinct_tags)):
            temp_prob = -1
            temp_key = 0
            for tags in distinct_tags:
                transmission_trigram = (distinct_tags[Prev_tag[i - 1][j]], distinct_tags[tag_t[i - 1][j]], tags)
                emmision_trigram = (distinct_tags[tag_t[i - 1][j]], tags, words[m])
                P_this_tag = P_t[i - 1][j] * probability_emmision(emmision_trigram) * probability_transmission(transmission_trigram)

                if P_this_tag > temp_prob:
                    temp_prob = P_this_tag

                    temp_key = tags
            Prev_tag[i].append(j)
            P_t[i].append(temp_prob)
            tag_t[i].append(distinct_tags.index(temp_key))
            
    
    return P_t, Prev_tag, tag_t
```

```python
## backward algo
def get_tags(a, b, c, length, k):
    ans = []
    l = len(a)
    ##to take top 50 
    temp = np.array(a[length])
    temp2 = temp.argsort()[-k:][::-1]
    for idx in temp2:
        ans1 = []
        ans1.append(distinct_tags[c[length][idx]])
        for i in range(1, l - 1):
            ans1.append(distinct_tags[c[length - i][idx]])
        ans.append(ans1)
    return ans
```

```python
Probs, Previous, Tags = viterbi("The Fulton County Grand")
```

```python
tags_of_sequence_top_k = get_tags(Probs, Previous, Tags, 4, 50)
```
